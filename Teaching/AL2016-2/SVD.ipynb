```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img = mpimg.imread('TinTin-Pequeno.jpg')
print(img.shape,img.dtype, img.nbytes/1000000)
```

```python
imggray = np.mean(img,-1).astype("uint8")
plt.figure(figsize=(7,10))
plt.imshow(imggray,cmap="gray");
print(imggray.nbytes/1000000)
```

```python
U, sigma, V = np.linalg.svd(imggray)
```

```python
n = 30
reconstimg = np.matrix(U[:, :n]) * np.diag(sigma[:n]) * np.matrix(V[:n, :])
plt.figure(figsize=(7,10))
plt.imshow(reconstimg,cmap="gray");
print( (U[:,:n].nbytes + sigma[:n].nbytes + V[:n,:].nbytes)/1000000 )
```
