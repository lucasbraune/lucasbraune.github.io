```python
import numpy as np
```

```python
A = np.array( [[-1,2,2],[-1,-4,-2],[-3,9,7]] )
#A = np.linalg.inv(A)
print(A)
```

```python
N = 10

v = np.array([1,0,0])

for i in range(0,N):
    vAntigo = v
    
    v = A.dot(v)    
    
    if vAntigo[0]!= 0:
        autovalor = v[0]/vAntigo[0]
        print(autovalor)
    else:
        print("Divisao por zero no autovalor")
        
    norma = np.sqrt( v.dot(v) ) 
    if norma != 0 :
        v = v/norma
        print(v)
    else:
        print("Divisao por zero no autovetor")
        break
```

```python
A = np.array( [[-1,2,2],[-1,-4,-2],[-3,9,7]] )
#A = np.random.rand(3,3)
print("Matriz original")
print(A)

N = 10 #iteracoes

for i in range(0,N):
    Q , R = np.linalg.qr(A)
    A = R.dot(Q)
    print("Iteracão:")
    print(A)
```

```python
#A = np.array( [[-1,2,2],[-1,-4,-2],[-3,9,7]] )
A = np.random.rand(3,3)
A = A.dot(np.transpose(A))
print("Matriz original (simetrica)")
print(A)

N = 3

S = np.identity(3)
for i in range(0,N):
    Q , R = np.linalg.qr(A)
    A = R.dot(Q)
    S = S.dot(Q)
    print("Iteracão:")
    print(A)
    
print("Autovalores:")
print(A)
print("Autovetores:")
print(S)
print("Matriz original:")
print(S.dot(A).dot(np.transpose(S)))
```
