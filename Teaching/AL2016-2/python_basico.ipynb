<h3>Importar módulos</h3>

Um <b>módulo</b> é um arquivo com extensão .py.<br>

Na mesma pasta deste caderno Jupyter está o arquivo fibonacci.py. Ele contém a função fib(n) que retorna o n-ésimo número de Fibonacci. (A sequência dos números de Fibonacci começa 1, 1, 2, 3, 5, 8, 13...)<br>

Para termos acesso ao código em fibonacci.py, escrevemos:

```python
import fibonacci
```

```python
fibonacci.fib(6)
```

```python
fibonacci.fib(6000)
```

Para importar a função fib diretamente, escrevemos

```python
from fibonacci import fib
```

```python
fib(10)
```

Se eu quisesse importar diretamente tudo que há no módulo fibonacci, eu escreveria

```python
from fibonacci import *
```
